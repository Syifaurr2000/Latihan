```python
title_map={"Capt":"Officer",
          "Col":"Officer",
          "Major":"Officer",
          "Johkheer":"Royalty",
          "Don":"Royalty",
          "Sir":"Royalty",
          "Dr":"Royalty",
          "Rev":"Officer",
          "The Countess":"Royalty",
          "Dona":"Royalty",
          "Mme":"Mrs",
          "Mlle":"Miss",
          "Ms":"Mrs",
          "Mr":"Mr",
          "Mrs":"Mrs",
          "Miss":"Miss",
          "Master":"Master",
          "Lady":"Royalty"}
data["Title"]=data.Title.map(title_map)
```

```python
data["Age"].isnull().sum()
```

```python
def passenger_type (row):
    if row['Age'] < 2:
        return 'Infant'
    elif (row['Age'] >= 2 and row['Age'] < 12):
        return 'Child'
    elif (row['Age'] >= 12 and row['Age'] < 18):
        return 'Youth'
    elif (row['Age'] >= 18 and row['Age'] < 65):
        return 'Adult'
    elif row['Age'] >= 65:
        return 'Senior'
    elif row['Title'] == 'Master':
        return 'Child'
    elif row['Title'] == 'Miss':
        return 'Child'
    elif row['Title'] == 'Mr' or row['Title'] == 'Mrs':
        return 'Adult'
    else:
        return 'Unknown'
```

```python
data['PassengerType'] = data.apply(lambda row: passenger_type(row),axis=1)
```

```python
x = data.drop('Survived', axis=1)
y = data.Survived
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=27)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN = KNeighborsClassifier().fit(x_train, y_train)
```

```python
KNN_pred = KNN.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, KNN_pred))
```

```python
pd.crosstab(y_test, KNN_pred, rownames=['Actual'], colnames=['Predicted'], margins=True)
```

```python
print(classification_report(y_test, KNN_pred))
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt
plt.rc("font", size=14)
```

```python
KNN_roc_auc = roc_auc_score(y_test, KNN.predict(x_test))
knn_fpr, knn_tpr, knn_thresholds = roc_curve(y_test, KNN.predict_proba(x_test)[:,1])
plt.figure()
plt.plot(knn_fpr, knn_tpr, label='KNN (AUC = %0.2f)' % KNN_roc_auc)
plt.plot([0, 1], [0, 1], 'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.show()
```
